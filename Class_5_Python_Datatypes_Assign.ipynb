```python
n = [34,23,89,56,2]
print(max(n))
print(min(n))
print(len(n))
print(sum(n))
print(type(n))
print("print greater than 10 : ",any(num > 10 for num in n)) # any 1 condition true output is True
print("print greater than 10 : ",all(num > 10 for num in n)) # all condition true output true
print("squares" , list(map(lambda n : n*2,n)))         # iterates all elements and give result
print("even nuber", list(filter(lambda n : n%2 ==0 ,n))) # it check only particular condition and provide results
```
